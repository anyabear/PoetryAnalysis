```python
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
poem_obj = open("poem_for_the_newly_damned.txt", "r")
just_poem = poem_obj.read()
print(just_poem)
```

```python
poem = TextBlob(poem_obj.read())
```

```python
sentences = poem.sentences
sentiment = poem.sentiment
print(poem)
```

```python
print(TextBlob(just_poem).sentiment.subjectivity)
```

```python
line_list = []
string = ""
for character in just_poem: 
    if character != '.':
        string = string + character
    if character == '.':
        line_list.append(string)
        string = ""
print(line_list)
```

```python
print("************ \n happy sentences \n************ \n ")
for a_thought in line_list: 
    print(TextBlob(a_thought).sentiment.polarity)
    print(a_thought)
print("\n************ \n sad sentences \n************ \n ")
for a_thought in sentences: 
    if TextBlob(a_thought).sentiment.polarity < 0.0:
        print(a_thought)
print("\n************ \n confusing sentences \n************ \n ")
for a_thought in poem.sentences: 
    if TextBlob(a_thought).sentiment.polarity == 0.0 and TextBlob(a_thought).sentiment.subjectivity != 0.0:
        print(a_thought)
```

```python
print("\n************ \n happy lines \n************ \n ")
for line in line_list:
    print(TextBlob(line).sentiment.polarity)
    print(line)
print("\n************ \n sad lines \n************ \n ")
for line in line_list:
    if TextBlob(line).sentiment.polarity < 0.0:
        print(line)
print("\n************ \n confusing lines \n************ \n ")
for line in line_list:
    if TextBlob(line).sentiment.polarity == 0.0 and TextBlob(line).sentiment.subjectivity != 0.0:
        print(line)
```

```python
print("\n************ \n happy words \n************ \n ")
for line in line_list:
    words = TextBlob(line).words
    for word in words:
        if TextBlob(word).sentiment.polarity > 0.0:
            print(word)
print("\n************ \n sad words \n************ \n ")
for line in line_list:
    words = TextBlob(line).words
    for word in words:
        if TextBlob(word).sentiment.polarity < 0.0:
            print(word)
print("\n************ \n confusing words \n************ \n ")
for line in line_list:
    words = TextBlob(line).words
    for word in words:
        if TextBlob(word).sentiment.polarity == 0.0 and TextBlob(word).sentiment.subjectivity != 0.0:
            print(word)
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
print(sid.polarity_scores(just_poem))
print("\n************ \n happy lines \n************ \n ")
for line in line_list:
    print(sid.polarity_scores(line)['pos'])
    print(line)
print("\n************ \n sad lines \n************ \n ")
for line in line_list:
    if sid.polarity_scores(line)['neg'] > 0.0:
        print(sid.polarity_scores(line)['neg'])
        print(line)
    # print(sid.polarity_scores(line))
```

```python
sentences = sent_tokenize(just_poem)
print(sentences)
```

```python
print("\n************ \nhappy sentences \n************ \n ")
for sent in sentences:
    
    if sid.polarity_scores(sent)['pos'] > 0.0:
        print(sid.polarity_scores(sent)['pos'])
        print(sent)
print("\n************ \nsad sentences\n************ \n ")
for sent in sentences:
    if sid.polarity_scores(sent)['neg'] > 0.0:
        print(sid.polarity_scores(sent)['neg'])
        print(sent)
    # print(sid.polarity_scores(line))
# for sent in sentences:
#     print(sent)
```
